```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
engine = create_engine("sqlite:///mutualfund_db.db")
```

```python
files = {
    "fund_master": "data/processed/cleaned_01_fund_master.csv",
    "nav_history": "data/processed/cleaned_02_nav_history_cleaned.csv",
    "aum_data": "data/processed/cleaned_03_aum_by_fund_house.csv",
    "monthly_sip": "data/processed/cleaned_04_monthly_sip_inflows.csv",
    "category_inflows": "data/processed/cleaned_05_category_inflows.csv",
     "industry folio": "data/processed/cleaned_06_industry_folio_count.csv",
    "scheme_performance": "data/processed/cleaned_07_scheme_performance.csv",
     "investor_transactions": "data/processed/cleaned_08_investor_transactions.csv",
    "portfolio_holdings": "data/processed/cleaned_09_portfolio_holdings.csv",
     "benchmark_indices": "data/processed/cleaned_10_benchmark_indices.csv",
}
for table_name, file_path in files.items():
    csv_count = len(pd.read_csv(file_path))
    query = f"SELECT COUNT(*) AS cnt FROM {table_name}"
    db_count = pd.read_sql(query, engine).iloc[0]["cnt"]
    status = "MATCH ✅" if csv_count == db_count else "MISMATCH ❌"
print(
        f"{table_name:<25} "
        f"CSV Rows: {csv_count:<10} "
        f"DB Rows: {db_count:<10} "
        f"{status}"
    )
```
